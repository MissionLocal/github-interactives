```python
df['race'] = df['race'].replace(races)
```

```python
df = df[['name', 'amount', 'committee_name', 'position', 'contest', 'slate', 'race']]
```

```python
# if the name and contest are the same, group by name and contest and sum the amount, keep all the other columns

df = df.groupby(['name', 'contest']).agg({'amount':'sum', 'committee_name':'first', 'position':'first', 'slate':'first', 'race':'first'}).reset_index()
```
