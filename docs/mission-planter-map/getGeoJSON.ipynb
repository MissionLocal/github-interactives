```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('planterInfo.csv')
```

```python
df
```

```python
geometry = gpd.points_from_xy(df['lng'], df['lat'])
geo_df = gpd.GeoDataFrame(df, geometry=geometry)
```

```python
geojson_file_path = 'coordinates.geojson'
geo_df.to_file(geojson_file_path, driver='GeoJSON')
```
